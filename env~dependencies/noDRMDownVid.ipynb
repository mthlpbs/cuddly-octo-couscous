#### **When Audio file have a audio file type**

````python
#@title <h6> **Fast method without reencode**
!ffmpeg -i "/content/drive/MyDrive/CacheFiles/v1.mp4" -i "/content/drive/MyDrive/CacheFiles/a1.mp3" -y -acodec copy -vcodec copy "/content/drive/MyDrive/{re1}.mp4"
!rm -rf "/content/drive/MyDrive/CacheFiles"
!rm -rf "/content/drive/bin/CacheFiles"
````

````python
#@title <h6> **Slow method With reencode Method**
!ffmpeg -i "/content/drive/MyDrive/CacheFiles/v1.mp4" -i "/content/drive/MyDrive/CacheFiles/a1.mp3" -c:v copy -c:a aac "/content/drive/MyDrive/{re1}.mp4"
!rm -rf "/content/drive/MyDrive/CacheFiles"
!rm -rf "/content/drive/bin/CacheFiles"
````
